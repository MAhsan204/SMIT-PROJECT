```python
import pandas as pd
import pyodbc
from sqlalchemy import create_engine

server = r'BTC-\SQLEXPRESS'
database = 'MISC'

conn_str = f'DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes;'


conn = pyodbc.connect(conn_str)
print("✅ Connected to SQL Server successfully!")

df = pd.read_csv('transformed_data.csv')
print("✅ CSV loaded into DataFrame. Sample:")
print(df.head())

engine = create_engine(f'mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server')

df.to_sql('YourTableName', con=engine, if_exists='replace', index=False)
print("✅ DataFrame uploaded successfully to SQL Server!")


df_from_sql = pd.read_sql("SELECT * FROM YourTableName", conn)
print("✅ Data read back from SQL Server:")
print(df_from_sql.head())
```
